```python
import pandas as pd
```

```python
file_path = "../data/HSA_HD_Final.csv"  
df = pd.read_csv(file_path, sep=',')
```

```python
df.head()
```

```python
# Xóa bỏ các cột không cần thiết: Mã ĐTN, Giới tính, loại ngoại ngữ (toàn N1 hết)
col_to_drops = ['Mã ĐTN', 'Giới tính', '10.Môn ngoại ngữ', '11.Môn ngoại ngữ']
# Xóa các cột không có giá trị tính toán (điểm các kì, học lực, hạnh kiểm), chỉ lấy điểm cả năm của các môn
for col in df.columns:
    can_drop = 'tổng kết CN' in col or 'Học lực CN' in col or 'Hạnh kiểm CN' in col or 'I' in col
    if can_drop:
        col_to_drops.append(col)
df = df.drop(col_to_drops, axis=1)
# Chỉ còn lại cột các điểm cả năm và điểm d1, d2, d3 và diem HSA
df.columns
```

```python
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])  # Chỉ in ra những cột có giá trị bị thiếu
```

```python
# Mỗi cột thiếu 1-2 dòng dữ liệu, ta có thể xóa các dòng dữ liệu này
df.dropna(inplace=True)
```

```python
print(df.info())  # Kiểm tra lại xem còn giá trị NaN không
```

```python
# Tính điểm trung bình của từng môn trong 3 năm
mon_hoc = ["Toán", "Văn", "Vật lí", "Hóa học", "Sinh học", "Lịch sử", "Địa lí", "GDCD", "Ngoại ngữ"]
for mon in mon_hoc:
    col_names = [f"10.{mon} CN", f"11.{mon} CN", f"12.{mon} CN"]
    # Công thức tính điểm trung bình: (L10 + L11 + L12 * 2) / 4 (ưu tiên điểm lớp 12)
    df[f"{mon}_TB"] = ((df[col_names[0]] + df[col_names[1]] + df[col_names[2]] * 2) / 4).round(2)
    # df[f"{mon}_TB"] = ((df[col_names[0]] + df[col_names[1]] + df[col_names[2]]) / 3).round(2)
    df.drop(col_names, axis=1, inplace=True)  # Xóa 3 cột cũ
df["KHXH_TB"] = df[["Lịch sử_TB", "Địa lí_TB"]].mean(axis=1).round(2)
df["KHTN_TB"] = df[["Vật lí_TB", "Hóa học_TB", "Sinh học_TB"]].mean(axis=1).round(2)
df.drop(["Lịch sử_TB", "Địa lí_TB", "GDCD_TB", "Vật lí_TB", "Hóa học_TB", "Sinh học_TB"], axis=1, inplace=True)
```

```python
df["KH_TB"] = df[["KHXH_TB", "KHTN_TB"]].mean(axis=1).round(2)
df.drop(["KHXH_TB", "KHTN_TB", 'Ngoại ngữ_TB'], axis=1, inplace=True)
df.to_csv("../data/HSA_HD_Final_Cleaned.csv", index=False)
print("Đã lưu file HSA_HD_Final_Cleaned.csv")
print(df.head())
```
